# Spectral Methods for Linear Advection

As an extremely simple example, we solve the linear advection equation
$$\frac{\partial u}{\partial t} = c \frac{\partial u}{\partial x}$$
using spectral methods. 

For simplicity, we study the equation on the domain 
$$\Omega=[0,2\pi]$$
with periodic boundary conditions.

We represent our function as a sum over positive and negative frequencies of Fourier modes for a total of $2N+1$ modes.
$$u(t,x) = \sum_{n=-N}^N u_n(t) e^{inx}$$
our vector of coefficients, then, is
$$\vec{u} = [u_{-N},u_{-N+1},...,u_0,u_{N}]$$

If we plug into the advection equation, we find that
$$\frac{d u_n}{dt} = i c n u_n.$$

Python can handle imaginary numbers no problem, so we don't have to worry about making everything real. Note that numpy uses the engineer's convention: $\sqrt{-1}=j$.

We can solve this numerically using the techniques we've learned so far.

Let's get to it. First, let's import our modules.

```python
# First we import our modules
%matplotlib inline
import numpy 
import matplotlib.pyplot as plt
from matplotlib import animation
from JSAnimation.IPython_display import display_animation
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

```python
# We define the ORDER of approximation 
N = 10
# And the speed of wave propogation
c = -1
# Here we define our basis functions
phi = lambda x,n: numpy.exp(1.j*n*x)
```

Let's define a function that maps a vector u_i to a function u(x). In other words, let's take a vector of coefficients and transform it into a real fucntion.

```python
def solution(u):
    """
    Takes a vector u of coefficients u_i and returns the true function
    u(x) = sum(u_i * e^(inx))
    """
    n = lambda i: i - N
    return lambda x: numpy.sum(numpy.array([u[i]*phi(x,n(i)) for i in range(len(u))]),axis=0)
```

Let's get it over with and define an initial function. We'll use 
$$u0=u(0,x) = \sin(e^{-(x-\pi)^2}).$$
Let's plot it:

```python
x = numpy.linspace(0,2*numpy.pi,100)
u_initial = lambda x: numpy.sin(numpy.exp(-(x-numpy.pi)**2))
plt.plot(x,u_initial(x))
```

How do we convert this into a spectral representation? We will want to take the inner product with every one of our functions. Because our basis functions aren't polynomials, we can't use fancy exact quadrature rules. Instead we use a very simplistic finite approximation of an integral. (We use the the rectangle rule.)

```python
dx = x[1]-x[0]
u0 = numpy.array([numpy.sum(numpy.array([dx * u_initial(x[i]+x[i+1])*phi(x[i]+x[i+1],-n)\
                                         for i in range(len(x)-1)]))\
                  for n in range(-N,N+1)])
```

Let's see if it looks right.

```python
plt.plot(x,numpy.real(solution(u0)(x))) # why do we have to use "real" here?
```

Okay, let's evolve this thing in time.

A word on the CFL condition. Here assume that $\Delta x$ is the wavelength of your smallest mode. In our case, this means that $\Delta x$ should be $N/2*pi$.

```python
tmax=2*numpy.pi
NUM_T=200
t = numpy.linspace(0,tmax,NUM_T)
dt=t[1]-t[0]
u = numpy.empty((len(t),len(u0)),dtype=complex) # why do we need to specify a data type here?
u[0] = u0 # set initial condition
n = numpy.arange(-N,N+1,dtype=float) # why do we need this?
# Integrate!
for i in range(1,len(t)): #RK2
    u_temp = u[i-1] + 0.5*dt*c*1.j*n*u[i-1]
    u[i] = u[i-1] + c*dt*1.j*n*u_temp
```

Let's animate it!

```python
fig = plt.figure()
ax = plt.axes(xlim=(0,2*numpy.pi),ylim=(0,3),xlabel=('x'),ylabel=('u'))
line, = ax.plot([],[],color='#003366', lw=2)
def animate(i):
    line.set_data(x,numpy.real(solution(u[i])(x)))
    return line
animation.FuncAnimation(fig, animate, frames=len(t), interval=10)
```
